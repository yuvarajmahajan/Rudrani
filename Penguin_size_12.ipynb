# label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_columns = ['species', 'island', 'sex']
```

```python
df['species'].unique()
```

```python
le_df = df.copy()
le = LabelEncoder()
for colm in categorical_columns:
    le_df[colm] = le.fit_transform(df[colm])
le_df.head()
```

```python
for col in categorical_columns:
    print(f"Unique {col}:{df[col].unique()}")
```

# one-hot-encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe_df = df.copy()
ohe = OneHotEncoder(sparse_output=False)
```

```python
df['species'].unique()
```

```python
ohe_df = ohe.fit_transform(df[categorical_columns])
ohe_df = pd.DataFrame(ohe_df, columns=ohe.get_feature_names_out(categorical_columns))
ohe_df = pd.concat([ohe_df,df])
```
